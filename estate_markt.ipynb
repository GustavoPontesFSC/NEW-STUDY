```python
A = a.array()
```

```python
x = A[:,0]
y = A[:,1]
print(x, y)
```

```python
Unique = np.unique(A[:,9])
ocean_proximity = []
for i in Unique:
    ocean_proximity.append(A[:,9] == i)
    
HV = A[:,8] > 400000
```

```python
for i in range(len(Unique)):
    print(Unique[i])
```

```python
for i in range(len(ocean_proximity)):
    plt.plot(x[ocean_proximity[i]] , y[ocean_proximity[i]] , '*', label = Unique[i])    
plt.plot(x[HV], y[HV], '*' ,color = 'deeppink', label = 'Value > U$400,000.00')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.legend()
plt.grid()
plt.show()
```
